Checking for outliers and dealing with them.

- Separating the different types of data

```python
## categorical data
Categorical_data = data.select_dtypes(include="object")
Continuous_data = data.select_dtypes(exclude="object")
```

```python
Categorical_data.head()
```

Drawing boxplots for the data

```python
def boxplot(dataframe,columnname):
    plt.figure(figsize=(15,15))
    for i,col in enumerate(columnname):
        plt.subplot(4,4,i+1)
        plt.boxplot(dataframe[col])
        plt.title(col)
    plt.show()
   
    
```

```python
Continuous_data.dtypes
```

```python
boxplot(Continuous_data,Continuous_data.columns)
```

Removing outliers

```python
def removing_outliers(Data):
    for i in Data.columns:
        if Data[i].dtype != "object":
            q1 = Data[i].quantile(0.25)
            q3 = Data[i].quantile(0.75)
            iqr = q3-q1
            lower_limit = q1-1.5*iqr
            upper_limit = q3+1.5*iqr
            Data[i] = np.where(Data[i]>upper_limit,upper_limit,Data[i])
            Data[i] = np.where(Data[i]<lower_limit,lower_limit,Data[i])
```

```python
removing_outliers(Continuous_data)
```

```python
boxplot(Continuous_data,Continuous_data.columns)
```

### New caeses
